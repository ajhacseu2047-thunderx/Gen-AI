```python
import torch
from diffusers import StableDiffusionPipeline
import os
from IPython.display import display # Added for visibility

# 1. Setup
model_id = "runwayml/stable-diffusion-v1-5"
output_dir = "synthetic_dataset"
os.makedirs(output_dir, exist_ok=True)

# 2. Load the pre-trained model [cite: 9]
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# 3. Input prompts (at least 5) [cite: 10]
prompts = [
    "A futuristic city skyline with flying cars at sunset",
    "A serene mountain landscape with a crystal clear lake",
    "A cute robot serving coffee in a vintage cafe",
    "An astronaut riding a horse on Mars, cinematic lighting",
    "A magical forest with glowing plants and mythical creatures"
]

# 4. Generate, Save, and Display [cite: 11, 12, 13]
for i, prompt in enumerate(prompts):
    print(f"--- Generating Image {i+1} ---")

    # Generate the image
    image = pipe(prompt).images[0]

    # Save the output
    file_path = os.path.join(output_dir, f"synthetic_image_{i+1}.png")
    image.save(file_path)

    # DISPLAY the output in the screen
    display(image)
```
